# Levantamiento de la base censo

```python
import pandas as pd 
import numpy as np 
pd.set_option('display.max_columns', None)
```

```python
Censo = pd.read_csv(r"Bases_Censo/conjunto_de_datos_iter_00CSV20.csv", encoding="UTF-8")
Censo.head()
```

Voy a reindexar dos municipio del estado de Chipas, hacerlo no afecta en el analisis. 

```python
Censo.loc[(Censo["ENTIDAD"] == 3) & (Censo["MUN"] == 8), "MUN"] = "4"
Censo.loc[(Censo["ENTIDAD"] == 3) & (Censo["MUN"] == 9), "MUN"] = "5"
```

Unicamente nos interesan los totales a nivel municipios

```python
Censo1 = Censo[(Censo["LOC"] == 0)&(Censo["MUN"] != 0)].copy()
Censo1[(Censo1["ENTIDAD"] == 7) & (Censo1["MUN"].isin([94,95,96]))]
```

Cambiamos los indices de la columna `MUN`, ya que el valor 95 no se encuentra, hacer esto no afecta el analisis. Solo es para hacer coincidir con nuestra variable de respuesta en `Votos_por_Municipio.csv`. 

```python
for i in range(95,126):
    Censo1.loc[(Censo1["ENTIDAD"] == 7) & (Censo1["MUN"] == i ), "MUN"] = i-1 
```

Creamos el identificador unico de cada municipio y corroboramos que el total de la poblacion coincide con lo reportado en la pagina del [INEGI](https://www.inegi.org.mx/temas/estructura/)

```python
Censo1["C_Municipio"] = (Censo1["ENTIDAD"].astype(str).str.zfill(2) +
                       "," +
                       Censo1["MUN"].astype(str).str.zfill(3))

print(f"Poblacin total:{Censo1["POBTOT"].sum()}")
```

Esto ultimo coincide con lo reportado en el [Censo 2020](https://es.wikipedia.org/wiki/M%C3%A9xico).

```python
Censo1.columns
```

Seleccionamos las variables mas relevantes de la base de datos, de acuerdo a `diccionario_datos_iter_00CSV20.csv` que acompaan la base de datos. 

```python
Columnas  = ["C_Municipio","NOM_ENT", "NOM_MUN","ENTIDAD", "MUN", 
                    "POBTOT", "P_12YMAS", "P_15YMAS",
                    "POB65_MAS", "PROM_HNV", "PHOG_IND", "P15YM_AN",
                    "GRAPROES", "PEA", "PDER_SS", "PAFIL_IPRIV",
                    "TVIVPARHAB", "PROM_OCUP", "VPH_DSADMA", "VPH_INTER",
                    "POBFEM", "POBMAS",
                    "P_18YMAS", "P_18YMAS_F", "P_18YMAS_M",
                    "P_18A24", "P_18A24_F", "P_18A24_M",
                    "P_25A29", "P_25A29_F", "P_25A29_M",
                    "P_30A34", "P_30A34_F", "P_30A34_M",
                    "P_35A39", "P_35A39_F", "P_35A39_M",
                    "P_40A44", "P_40A44_F", "P_40A44_M",
                    "P_45A49", "P_45A49_F", "P_45A49_M",
                    "P_50A54", "P_50A54_F", "P_50A54_M",
                    "P_55A59", "P_55A59_F", "P_55A59_M",
                    "P_60YMAS", "P_60YMAS_F", "P_60YMAS_M",
                    "PCON_DISC", "PE_INAC", "PSINDER"]

Censo2 = Censo1[Columnas].reset_index(drop=True)

# Verificamos el tipo de dato de las columas
Censo2.dtypes
```

Tenemos variables numericas que tienen formato de objeto

```python
Numericas = [x for x in Columnas if x not in ["C_Municipio","NOM_ENT","NOM_MUN","ENTIDAD","MUN"]]

for col in Numericas:
    Censo2[col] = pd.to_numeric(Censo2[col], errors="coerce").astype(float)

Censo2.head()
```

```python
Censo2.dtypes
```

La escala de las variables es distinta, por ello conviene normalizar 

```python
Censo2["POBMAS"] = 100*(Censo2["POBMAS"]/Censo2["POBTOT"])
Censo2["POBFEM"] = 100*(Censo2["POBFEM"]/Censo2["POBTOT"]) 

Censo2["P_18A29"] = 100*(Censo2["P_18A24"]+ Censo2["P_25A29"])/Censo2["POBTOT"]
Censo2["P_18A29_F"] = 100*(Censo2["P_18A24_F"]+ Censo2["P_25A29_F"])/Censo2["POBTOT"]
Censo2["P_18A29_M"] = 100*(Censo2["P_18A24_M"]+ Censo2["P_25A29_M"])/Censo2["POBTOT"]

Censo2["P_30A44"] = 100*(Censo2["P_30A34"]+ Censo2["P_40A44"])/Censo2["POBTOT"]
Censo2["P_30A44_F"] = 100*(Censo2["P_30A34_F"]+ Censo2["P_40A44_F"])/Censo2["POBTOT"]
Censo2["P_30A44_M"] = 100*(Censo2["P_30A34_M"]+ Censo2["P_40A44_M"])/Censo2["POBTOT"]

Censo2["P_45A59"] = 100*(Censo2["P_45A49"]+ Censo2["P_50A54"]+  Censo2["P_55A59"])/Censo2["POBTOT"]
Censo2["P_45A59_F"] = 100*(Censo2["P_45A49_F"]+ Censo2["P_50A54_F"]+  Censo2["P_55A59_F"])/Censo2["POBTOT"]
Censo2["P_45A59_M"] = 100*(Censo2["P_45A49_M"]+ Censo2["P_50A54_M"]+  Censo2["P_55A59_M"])/Censo2["POBTOT"]

Censo2["P_60YMAS"] = 100*(Censo2["P_60YMAS"]) / Censo2["POBTOT"]
Censo2["P_60YMAS_F"] = 100*(Censo2["P_60YMAS_F"]) / Censo2["POBTOT"]
Censo2["P_60YMAS_M"] = 100*(Censo2["P_60YMAS_M"]) / Censo2["POBTOT"]

Censo2["POB65_MAS"] = 100*(Censo2["POB65_MAS"]) / Censo2["POBTOT"]

Censo2["P_DISC"] = 100*(Censo2["PCON_DISC"]) / Censo2["POBTOT"]

Censo2["PEA"] = 100*(Censo2["PEA"]) / Censo2["P_12YMAS"]

Censo2["ANALF"] = 100*(Censo2["P15YM_AN"]) / Censo2["P_15YMAS"]

Censo2["PHOG_IND"] = 100*(Censo2["PHOG_IND"]) / Censo2["POBTOT"]
Censo2["PROM_HNV"] = 100*(Censo2["PROM_HNV"]) 
Censo2["GRAPROES"] = 100*(Censo2["GRAPROES"]) 


Censo2["PDER_SS"] = 100*(Censo2["PDER_SS"]) / Censo2["POBTOT"]
Censo2["PDER_SINDER"] = 100*(Censo2["PSINDER"]) / Censo2["POBTOT"]
Censo2["PAFIL_IPRIV"] = 100*(Censo2["PAFIL_IPRIV"]) / Censo2["POBTOT"]


Censo2["VPH_INTER"] = 100*(Censo2["VPH_INTER"]) / Censo2["TVIVPARHAB"]
Censo2["VPH_DSADMA"] = 100*(Censo2["VPH_DSADMA"]) / Censo2["TVIVPARHAB"]

columnas_Base = ["C_Municipio","NOM_ENT","ENTIDAD","NOM_MUN","MUN",
                 "POBMAS","POBFEM","P_18A29","P_18A29_F","P_18A29_M",
                 "P_30A44","P_30A44_M","P_30A44_F","P_45A59","P_45A59_F",
                 "P_45A59_M","P_60YMAS","P_60YMAS_M","P_60YMAS_F","POB65_MAS",
                 "P_DISC","PEA","ANALF","PHOG_IND","PROM_HNV","GRAPROES",
                 "PDER_SS","PDER_SINDER","PAFIL_IPRIV","VPH_INTER","VPH_DSADMA"]

Base_censo = Censo2[columnas_Base]
```
