# Levantamiento de la base censo

```python
import pandas as pd 
import numpy as np 
pd.set_option('display.max_columns', None)
```

```python
Censo = pd.read_csv(r"Bases_Censo/Censo_INE2020.gz", compression="gzip", encoding="utf-8")
Censo.head()
```

Voy a reindexar dos municipio del estado de Chipas, hacerlo no afecta en el analisis. 

```python
Censo.loc[(Censo["ENTIDAD"] == 3) & (Censo["MUN"] == 8), "MUN"] = "4"
Censo.loc[(Censo["ENTIDAD"] == 3) & (Censo["MUN"] == 9), "MUN"] = "5"
```

Unicamente nos interesan los totales a nivel municipios

```python
Censo1 = Censo[(Censo["LOC"] == 0)&(Censo["MUN"] != 0)].copy()
Censo1[(Censo1["ENTIDAD"] == 7) & (Censo1["MUN"].isin([94,95,96]))]
```

Cambiamos los indices de la columna `MUN`, ya que el valor 95 no se encuentra, hacer esto no afecta el analisis. Solo es para hacer coincidir con nuestra variable de respuesta en `Votos_por_Municipio.csv`. 

```python
for i in range(95,126):
    Censo1.loc[(Censo1["ENTIDAD"] == 7) & (Censo1["MUN"] == i ), "MUN"] = i-1 
```

Creamos el identificador unico de cada municipio y corroboramos que el total de la poblacion coincide con lo reportado en la pagina del [INEGI](https://www.inegi.org.mx/temas/estructura/)

```python
Censo1["C_Municipio"] = (Censo1["ENTIDAD"].astype(str).str.zfill(2) +
                       "," +
                       Censo1["MUN"].astype(str).str.zfill(3))

print(f"Poblacin total:{Censo1["POBTOT"].sum()}")
```

Esto ultimo coincide con lo reportado en el [Censo 2020](https://es.wikipedia.org/wiki/M%C3%A9xico).
